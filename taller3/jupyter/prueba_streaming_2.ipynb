```python
from flurs.recommender.user_knn import UserKNNRecommender
from converter.converter import Converter
from flurs.data.entity import User, Item, Event
import numpy as np

'''
[Common]
Dataset: ML100k
Trial: 1

[Model]
Name: user-knn
Epoch: 1
'''
dataset = "ML100k"
k =  5


data = Converter().convert(dataset=dataset)

print (data.n_user)
print (data.n_item)
print (data.n_sample)
print (data.n_batch_train)
print (data.n_batch_test)
print (data.n_test)
print (len(data.ratings))
print (data.ratings[0])

print (data.ratings)
#list of events
print (len(data.samples))

#print (data.dts)




from flurs.recommender.mf import MFRecommender

recommender = MFRecommender(k=40)
#recommender = UserKNNRecommender(k=k)
#rec.initialize()
recommender.initialize()

items_indexes = []
evt_window01 = data.samples[:500]
for evt in evt_window01:
    print ("***")

    if recommender.is_new_user(evt.user.index):
        recommender.register_user(evt.user)
        #print (evt.user)
        print("new user")
        print (evt.user.index)

    if recommender.is_new_item(evt.item.index):
        recommender.register_item(evt.item)
        items_indexes.append(evt.item.index)
        #print (evt.item)
        print("new item")
        print (evt.item.index)

    #print (evt.value)
    #print (evt.context)
    recommender.update(evt)

print (recommender.users)
#print (recommender.items)
```

## listado de candidados

```python
from numpy  import array
candidates= array( items_indexes )
print("candidates")
print (candidates)








#  model, res = self.__run(create)
```
